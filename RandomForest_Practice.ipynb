```
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from google.colab import files
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
```

```
df = pd.read_csv('data_cleaned.csv')
```

```
df.shape
```

```
df.head()
```

### Split features into Target and Independent Features

```
x = df.drop('Survived', axis = 1)
y = df.Survived
```

### Split the Dataset into Train and Test Dataset

```
train_x, test_x, train_y, test_y = tts(x,y, random_state = 123, stratify = y)
```

```
train_x.shape, test_x.shape
```

### Random Forest Classifier Implementation

```
RF = RandomForestClassifier(random_state=123)
```

```
RF.fit(train_x,train_y)
```

```
rf_train_pred = RF.predict(train_x)
```

```
RF.score(train_x,train_y)
```

```
train_acc = accuracy_score(rf_train_pred, train_y)
train_acc
```

```
rf_test_pred = RF.predict(test_x)
```

```
RF.score(test_x,test_y)
```

```
test_acc = accuracy_score(rf_test_pred, test_y)
test_acc
```

### Tweak the Hyperparameters of the Random Forest Model

```
RF2 = RandomForestClassifier(n_estimators=200,
                             max_depth =5,
                             min_samples_split = 25,
                             min_samples_leaf = 25)
```

```
RF2.fit(train_x,train_y)
```

```
RF2.score(train_x,train_y)
```

```
RF2.score(test_x,test_y)
```

```
n_trees = [50,100,150,200]
max_feat = [3,5,7]
max_dep = [3,4,5,6]
min_samples_split = [10,20,30,40]
```

```
acc_score = []
trees = []
dep = []
min_split =[]
max_features = []

def rf_iterate(n_trees, max_feat, max_dep, min_samples_split):
  
  for i in n_trees:
    for j in max_feat:
      for k in max_dep:
        for l in min_samples_split:
          RF3 = RandomForestClassifier(n_estimators=i,
                               max_depth = k,
                               max_features = j,
                               min_samples_split = l,
                               random_state = 123)
          RF3.fit(train_x,train_y)
          acc_score.append(RF3.score(test_x,test_y))
          trees.append(i)
          dep.append(k)
          min_split.append(l)
          max_features.append(j)
```

```
rf_iterate(n_trees, max_feat,max_dep,min_samples_split)
```

```
perf_df = pd.DataFrame({'Trees':trees,
                        'Depth' : dep,
                        'Min_Sam_Spl' : min_split,
                        'Max_Features_spl' :max_features,
                        'Accuracy' : acc_score
    
})
```

```
perf_df[perf_df.Accuracy==perf_df.Accuracy.max()]
```

So we can select a model with either of the above 2 sets of parameters and improve the accuracy score of the model to 0.825 as opposed to 0.78 or 0.80 with earlier models.
